Lây dataset từ roboflow và khai báo thư viện để lưu kết quả về drive

```

!pip install roboflow
import os
import wandb
from google.colab import userdata
from roboflow import Roboflow

# Lấy API Key từ mục Secrets (hình chiếc chìa khóa) trên Colab để bảo mật
roboflow_api_key = userdata.get('ROBOFLOW_API_KEY')

wandb.login()
wandb.init(project="Cheating_Detection_Pro", job_type="training")
rf = Roboflow(api_key=roboflow_api_key)
project = rf.workspace("tan-thong").project("deepnew-rmes5")
version = project.version(3)
dataset = version.download("yolov8")
```

Tiền xử lý và train

```
from ultralytics import YOLO
import os

new_project_path = "/content/drive/MyDrive/YOLOv8_Deep"
os.makedirs(new_project_path, exist_ok=True)

model = YOLO("yolov8m.pt")

results = model.train(
    data=f"{dataset.location}/data.yaml",
    epochs=40,
    imgsz=640,
    batch=16,
    device=0,

    optimizer="auto",
    patience=30,
    amp=False,

    mosaic=1.0,
    mixup=0.2,
    degrees=15.0,
    hsv_s=0.7,
    hsv_v=0.4,
    fliplr=0.5,
    scale=0.5,

    project=new_project_path,
    name="Train_Lan_1",
    save=True,
    plots=True
)
```

Tiếp tục train khi bị ngừng
